```python
import random
import smtplib

# Function to generate a 6-digit OTP
def generate_otp():
    return ''.join([str(random.randint(0,9)) for i in range(6)])

# Function to send OTP to the user's email address
def send_otp_email(email, otp):
    sender_email = input("Enter sender's email: ")
    sender_password = input("Enter sender's password: ")

    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(sender_email, sender_password)

    msg = 'Hello, Your OTP is ' + str(otp)
    server.sendmail(sender_email, email, msg)
    server.quit()

# Function to prompt the user to enter the OTP received in their email
def prompt_for_otp():
    return input("Enter the OTP received in your email: ")

# Function to verify if the entered OTP matches the generated OTP
def verify_otp(otp, entered_otp):
    return otp == entered_otp

# Main function
def main():
    otp = generate_otp()
    recipient = input("Enter your email address: ")
    send_otp_email(recipient, otp)  # Replace with the recipient's email address

    # Allowing three attempts to enter OTP
    attempts = 3
    while attempts > 0:
        entered_otp = prompt_for_otp()
        if verify_otp(otp, entered_otp):
            print("OTP verified. Access granted!")
            break
        else:
            attempts -= 1
            if attempts > 0:
                print("Incorrect OTP. Please try again. Attempts left:", attempts)
            else:
                print("Maximum attempts reached. Access denied.")

if __name__ == "__main__":
    main()
```
